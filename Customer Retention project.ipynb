```python
def value_count(col):
    diff_count=len(data[col].value_counts())
    if diff_count<5:
        plt.figure(figsize=(10,6))
    elif diff_count<10:
        plt.figure(figsize=(10,6))
        plt.xticks(rotation=90)
    elif diff_count<20:
        plt.figure(figsize=(25,6))
        plt.xticks(rotation=90)
    else:
        plt.figure(figsize=(20,6))
        plt.xticks(rotation=90)
    sns.countplot(x=col,data=data,orient='v',palette='Accent')
    sns.countplot(data[col],hue=data['gender'],palette='Dark2_r')
    plt.show()
    
    #column infomation
    print("uniquevalues \n", data[col].unique())
    print("value counts \n",data[col].value_counts())
    print("\n")
    # Let's check the percentage of data classification in each string attribute
    print('percentage of unique values')
    print(round(data[col].value_counts()/269*100),2)
    print("---------------------------")
```

above defined function is for plot all the columns 

in plot we have 2 types of plots one indicates count of variables and in that only it gives a shade of hue, gender is using as hue
